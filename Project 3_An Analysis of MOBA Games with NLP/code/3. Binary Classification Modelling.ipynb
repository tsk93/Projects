Before training our data, we will add 2 more data pre-processing steps.

- Lemmatization
- Stemming

---
#### Lemmatization

Lemmatization is the process of grouping together the different inflected forms of a word so they can be analyzed as a single item. Lemmatization is similar to stemming but it brings context to the words, so it links words with similar meanings to one word. 

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_data(x):
    words=[]
    for i in x.split(' '):
        words.append(lemmatizer.lemmatize(i))
    return words
```

```python
df['post_EDA_lemmatized']=df['cleaned_tokenised_post_no_punct_no_stopwords'].map(lambda x: lemmatize_data(x))
```

```python
df['post_EDA_lemmatized']
```

---
#### Stemming

Stemming is a natural language processing (NLP) technique that lowers inflection in words to their root forms, hence aiding in the preprocessing of text, words, and documents for text normalization. This reduces words to their basic form or stem, which may or may not be a legitimate word in the language.

```python
p_stemmer = PorterStemmer()
```

```python
def stem_data(x):
    words=[]
    for i in x:
        words.append(p_stemmer.stem(i))
    return ' '.join(words)
```

```python
df['post_EDA_lemmatized_stemmed']=df['post_EDA_lemmatized'].map(lambda x: stem_data(x))
```

First we initialize the vectorizer to convert the train data into a bag-of-words model, where each row of numbers denotes the frequency of different words in a particular post.

```python
#create vectorizer to transform training text data into numerical input

cvec = CountVectorizer(stop_words=all_stopwords)
cvec.fit(X_train)
X_train_cvec = cvec.transform(X_train)
X_test_cvec = cvec.transform(X_test)
X_train_df = pd.DataFrame(X_train_cvec.toarray(),columns=cvec.get_feature_names_out())
X_train_df.head()
```

Classification problems are supervised learning problems where the training dataset consists of independent variables and response variables (label). These problems are generally trained on a few models as listed below, and we will explore each of them in this notebook.

- Naive Bayes
- Logistic Regression
- k-Nearest Neighbors
- Random Forest

For each model, we start with a wide range for fields of interest, and narrow around the optimally selected value and gauge the degree of accuracy increase (or decrease). Through trial and error, we are able to select hyperparameters that will return the most optimal modelling results. We will write a function as well to print out the classification metrics after a model has been trained, as shown below.

```python
def classification_metrics(classifier,gridsearch,X_test,y_test):
    print(f'{classifier}'.center(60, '-'))
    print('\n')
    print(f'Gridsearch Best Score: {gridsearch.best_score_:.4f}')
    print(f'Train Score: {gridsearch.score(X_train, y_train):.4f}')
    print(f'Test Score: {gridsearch.score(X_test, y_test):.4f}')
    print(f'Hyperparameters: {gridsearch.best_params_}')
    print('\n')

    pred = gridsearch.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
    print("True Negative: %s" % tn)
    print("False Positive: %s" % fp)
    print("False Negative: %s" % fn)
    print("True Positive: %s" % tp)
    print(f"\nAccuracy: {round(((tn + tp) / (tn + fp + fn + tp)),4)}")
    print(f"Sensitivity: {round((tp / (tp + fn)),4)}")
    print(f"Specificity: {round((tn / (tn + fp)),4)}")
    print(f"Precision: {round((tp / (tp + fp)) ,4)}")
    print('\n')

    fig,axes=plt.subplots(1,2,figsize=(20,6))
    sns.set(font_scale = 1)

    sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt='d',cmap='Blues',ax=axes[0])

    axes[0].set_xlabel("Predicted Label", fontsize=14, labelpad=10)
    axes[0].xaxis.set_ticklabels(['League of Legends', 'Dota 2'])

    axes[0].set_ylabel("Actual Label", fontsize=14, labelpad=10)
    axes[0].yaxis.set_ticklabels(['League of Legends', 'Dota 2'])

    axes[0].set_title(f"Confusion Matrix derived from {classifier}", fontsize=14, pad=10)
    
    plot_roc_curve(gridsearch, X_test, y_test,ax=axes[1])
    axes[1].set_xlabel("True Negative Rate", fontsize=14, labelpad=10)
    axes[1].set_ylabel("True Positive Rate", fontsize=14, labelpad=10)
    axes[1].set_title(f"Receiver Operating Characteristic (ROC) Curve derived from {classifier}", fontsize=14, pad=10)
    plt.plot([0, 1], [0, 1],label='baseline', linestyle='--')
    plt.show()
    print('\n')
    print('Classification Report'.center(60,'-'))
    print(classification_report(y_test, pred,target_names=['League of Legends', 'Dota 2']))
```

### Model 1: Naive Bayes (Multinomial NB)

The Naive Bayes classification algorithm is a technique that relies on Bayes Theorem and a key assumption is that our features are independent of one another. The advantage of making such an assumption makes it easier to calculate probabilities but it might be unrealistic, especially in the case of text data. While our classifications might be accurate, our predicted probabilities might not be reliable. In particular, we will be using the Multinomial NB model for this project as it is compatible with discrete features (eg. word count in text classification).

```python
# Model 1
pipe_1 = Pipeline([('cvec', CountVectorizer(stop_words=all_stopwords)),('mnb', MultinomialNB())])

pipe_1_params = {
    'cvec__max_features' : range(2800,4000,50),
    'cvec__min_df' : [2,3],
    'cvec__max_df' : [0.8,0.9,0.95],
    'cvec__ngram_range' : [(1,1),(1,2),(1,3)]
}

mnb = GridSearchCV(pipe_1,param_grid=pipe_1_params,cv=5,n_jobs=-1)
```

```python
mnb.fit(X_train, y_train)
```

```python
mnb.best_params_
```

```python
classification_metrics('Multinomial Naive Bayes',mnb,X_test,y_test)
```

### Model 2: Logistic Regression

Logistic regression is commonly used for binary classification algorithms and it estimates the probability of an event occurring based on a given dataset of independent variables. Numerical encoding (0/1) is commonly done to convert categorical variables into numerical values so that it could be easily fitted to a machine learning model. It will predict continuous probabilities between 0 and 1.

```python
pipe_2 = Pipeline([('cvec', CountVectorizer(stop_words=all_stopwords)),('logreg', LogisticRegression(solver='saga'))])

pipe_2_params = {
    'cvec__max_features' : list(range(2000,4001,200)),
    'cvec__min_df' : [2,3,4],
    'cvec__max_df': [0.8,0.9,0.95],
    'cvec__ngram_range' : [(1,1),(1,2)],
    'logreg__penalty': ['l1', 'l2', 'elasticnet'],
    'logreg__C': [10**x for x in [0,1,2,3]]
}

lr = GridSearchCV(pipe_2,param_grid=pipe_2_params,cv=5,n_jobs=-1)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.best_params_
```

```python
classification_metrics('Logistic Regression',lr,X_test,y_test)
```

### Model 3: K-Nearest Neighbors (KNN)

A KNN classifier works by finding the k nearest points in the training dataset and uses their class to predict the class or value of a new data point. As KNN is a majority voting algorithm, it is generally preferred to use an odd value of k. This is because an even value of k may lead to a tie in samples at the decision boudaries.

```python
pipe_3 = Pipeline([('cvec', CountVectorizer(stop_words=all_stopwords)),('knn', KNeighborsClassifier())])

pipe_3_params = {
    'cvec__max_features' : [50,100,150,200],
    'cvec__min_df' : [2,3,4,5],
    'cvec__max_df': [0.8,0.9,0.95],
    'cvec__ngram_range' : [(1,1),(1,2),(1,3)],
    'knn__n_neighbors': [3,5,7,9,11,13]
}

knn = GridSearchCV(pipe_3,param_grid=pipe_3_params,cv=5,n_jobs=-1)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.best_params_
```

```python
classification_metrics('K-Nearest Neighbours',knn,X_test,y_test)
```

### Model 4: Random Forest

Random forest is a supervised learning algorithm. The forest it builds is an ensemble of decision trees, usually trained with the bagging method. The general idea of the bagging method is that a combination of learning models increases the overall result. Multiple decision trees are built standalone and merged together eventually to get a more accurate and stable prediction. Each tree generally takes on a random subset of features for node splitting, which makes them less correlated and therefore strengthening the model.

```python
pipe_4 = Pipeline([('cvec', CountVectorizer(stop_words=all_stopwords)),('rf', RandomForestClassifier())])

pipe_4_params = {
    'cvec__max_features' : list(range(500,7001,500)),
    'cvec__min_df' : [2,3,4],
    'cvec__max_df': [0.8,0.9,0.95],
    'cvec__ngram_range' : [(1,1),(1,2),(1,3)]
}

rf = GridSearchCV(pipe_4,param_grid=pipe_4_params,cv=5,n_jobs=-1)
```

```python
rf.fit(X_train,y_train)
```

```python
classification_metrics('Random Forest',rf,X_test,y_test)
```

```python
model = [mnb, lr, knn, rf]
name = ['Multinomial Naive Bayes','Logistic Regression', 'K-Nearest Neighbours', 'Random Forest']
train_score=[]
test_score=[]
sensitivity=[]
specificity=[]
precision=[]
f1_score=[]
roc_auc=[]
t_p=[]
t_n=[]
f_p=[]
f_n=[]

for i in model:
    pred=i.predict(X_test)
    train_score.append(round(i.score(X_train, y_train),4))
    test_score.append(round(i.score(X_test, y_test),4))
    tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
    t_p.append(tp)
    t_n.append(tn)
    f_p.append(fp)
    f_n.append(fn)
    sensitivity.append(round((tp / (tp + fn)),4))
    specificity.append(round((tn / (tn + fp)),4))
    precision.append(round((tp / (tp + fp)),4))
    f1_score.append(round((tp / (tp+0.5*(fp + fn))),4))
    roc_auc.append(round(roc_auc_score(y_test, i.predict_proba(X_test)[:,1]),4))
    
final_dict = dict()
final_dict['Model'] = name
final_dict['Training Score'] = train_score
final_dict['Testing Score'] = test_score
final_dict['TP'] = t_p
final_dict['TN'] = t_n
final_dict['FP'] = f_p
final_dict['FN'] = f_n
final_dict['Sensitivity'] = sensitivity
final_dict['Specificity'] = specificity
final_dict['Precision'] = precision
final_dict['F1 Score'] = f1_score
final_dict['ROC-AUC Score'] = roc_auc
final_model=pd.DataFrame(final_dict).sort_values(by='Testing Score',ascending=False).reset_index(drop=True)
final_model
```
